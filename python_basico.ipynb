#Python Básico para Análise de Dados

Este notebook contém meus primeiros estudos relacionados à aplicação de Python à análise de dados.

#Tipos de dados

Strings (str): texto

Number (int): número inteiro

Float: número real com fração

Complex: número complexo. Exemplo: x + yj = z

Bool: verdadeiro ou falso

#Exemplos práticos

```python
media = 8
a1 = a2 = a3 = 0.7
estado = False

#Duas variáveis simultaneamente
nome, idade = 'Gabriella', 22

#Função type

print(type(media))
print(type(nome))
print(type(estado))
print(type(1+2j))

#Operação

a = 2
b = 10
c = a*b
print(c)
print(type(a1*idade))

#Função isinstance - indica T or F
print(isinstance(media, int))
print(isinstance(nome, bool))
##Função isinstance OU
print(isinstance(a1, (complex, float)))


```
